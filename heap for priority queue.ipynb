```python
class Max_Heap:
    def __init__(self, L):
        self.size = len(L)
        self.heap = L.copy()
    def build_max_heap(self): # n/4 (1 c)+n/8 (2 c)+n/16 (3 c)+…+ 1 (lg n c) => O(N), 1c means 1 level away from leaves
        i = self.size//2 
        while i>=0:
            self.max_heapify(i) 
            i -= 1
    def max_heapify(self, i): # O(logN)
        parent = self.heap[i]
        left_child_index = 2*i + 1
        right_child_index = 2*i + 2
        left_child = self.heap[left_child_index] if left_child_index < self.size else -math.inf
        right_child = self.heap[right_child_index] if right_child_index < self.size else -math.inf
        if left_child > parent and left_child >= right_child:
            self.heap[i] = left_child
            self.heap[left_child_index] = parent
            self.max_heapify(left_child_index)
        elif right_child > parent and right_child >= left_child:
            self.heap[i] = right_child
            self.heap[right_child_index] = parent
            self.max_heapify(right_child_index)
        else:
            pass
    def calculate_parent_index(self, index):
        if index%2 == 0:
            parent_index = index // 2 - 1
        else:
            parent_index = index // 2
        return parent_index
    def insert(self, ele):
        if len(self.heap) > self.size:
            self.heap[self.size] = ele
        else:
            self.heap.append(ele)
        self.size += 1
        son_index = self.size - 1
        parent_index = self.calculate_parent_index(son_index)
        while parent_index >= 0:
            if ele > self.heap[parent_index]:
                temp = self.heap[parent_index]
                self.heap[parent_index] = ele
                self.heap[son_index] = temp
                son_index = parent_index
                parent_index = self.calculate_parent_index(son_index)
            else:
                break
    def extract_max(self):
        if self.size:
            max_ele = self.heap[0]
            self.heap[0] = self.heap[self.size-1]
            self.size -= 1
            self.max_heapify(0)
            return max_ele
    def heapsort(self):
        sorted_list = []
        while self.size:
            sorted_list.append(self.extract_max())
        return sorted_list
```

```python
class Min_Heap:
    def __init__(self, L):
        self.size = len(L)
        self.heap = L.copy()
    def build_min_heap(self): # n/4 (1 c)+n/8 (2 c)+n/16 (3 c)+…+ 1 (lg n c) => O(N), 1c means 1 level away from leaves
        i = self.size//2 
        while i>=0:
            self.min_heapify(i) 
            i -= 1
    def min_heapify(self, i): # O(logN)
        parent = self.heap[i]
        left_child_index = 2*i + 1
        right_child_index = 2*i + 2
        left_child = self.heap[left_child_index] if left_child_index < self.size else math.inf
        right_child = self.heap[right_child_index] if right_child_index < self.size else math.inf
        if left_child < parent and left_child <= right_child:
            self.heap[i] = left_child
            self.heap[left_child_index] = parent
            self.min_heapify(left_child_index)
        elif right_child < parent and right_child <= left_child:
            self.heap[i] = right_child
            self.heap[right_child_index] = parent
            self.min_heapify(right_child_index)
        else:
            pass
    def calculate_parent_index(self, index):
        if index%2 == 0: # do not forget
            parent_index = index // 2 - 1
        else:
            parent_index = index // 2 
        return parent_index
    def insert(self, ele):
        self.heap.append(ele)
        self.size += 1
        son_index = self.size - 1
        parent_index = self.calculate_parent_index(son_index)
        while parent_index >= 0:
            if ele < self.heap[parent_index]:
                temp = self.heap[parent_index]
                self.heap[parent_index] = ele
                self.heap[son_index] = temp
                son_index = parent_index
                parent_index = self.calculate_parent_index(parent_index)
            else:
                break
    def extract_min(self):
        if self.size:
            min_ele = self.heap[0]
            self.heap[0] = self.heap[self.size-1]
            self.size -= 1
            self.min_heapify(0)
            return min_ele
    def heapsort(self):
        sorted_list = []
        while self.size:
            sorted_list.append(self.extract_min())
        return sorted_list
```

```python
test1 = []
for i in range(100):
    test1.append(random.random())
test2 = test1.copy()
test3 = test1.copy()
max_heap = Max_Heap(test1)
max_heap.build_max_heap()
min_heap = Min_Heap(test1)
min_heap.build_min_heap()
for i in range(100):
    r = random.random()
    max_heap.insert(r)
    min_heap.insert(r)
    test2.append(r)
    test3.append(r)
test2.sort()
test3.sort(reverse=True)
assert(max_heap.heapsort()== test3)
assert(min_heap.heapsort() == test2)
```
